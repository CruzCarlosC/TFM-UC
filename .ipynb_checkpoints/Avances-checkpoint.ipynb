### Track.py

Caracteristicas de la trayectoria de la particular
Toma como base las soluciones de la ecuacion de una particula relativista en un campo uniforme<br>A partir de 
$$\frac{dp^{\alpha}}{d\tau} = q F^{\alpha \beta} u_{\beta}$$ <br>Se llega a 
$$\frac{d\vec{p}}{d\tau} = kq \vec{v} x \vec{B}$$<br>Con $\vec{p} = \gamma m \vec{u}$ y k una constante en funcion de las unidades $k=0.29979 \frac{GeV}{cTm}$ y $w=\frac{qB}{mc\gamma}$<br><br>
$$x=\frac{p}{w}(cos(wt) - cos(\varphi))$$ <br>
$$y=\frac{p}{w}(sin(wt) - sin(\varphi))$$ <br>
Y z puede escribirse en terminos de $\eta$ <br>
$$z=\frac{p sinh(\eta)}{m \gamma} t$$
- Usa caracteristicas de un electron en un campo B=4 T
